# 项目：评估和清理英国电商公司销售数据

## 分析目标

此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 简介

原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。

数据每列的含义如下：
- `InvoiceNo`: 发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。
- `StockCode`: 产品代码。5位数，作为产品的唯一标识符。
- `Description`: 产品名称。
- `Quantity`: 产品在交易中的数量。
- `InvoiceDate`: 发票日期和时间。交易发生的日期和时间。
- `UnitPrice`: 单价。价格单位为英镑（£）。
- `CustomerID`: 客户编号。5位数，作为客户的唯一标识符。
- `Country`: 国家名称。客户所居住的国家的名称。

## 清理数据

根据前面评估部分得到的结论,我们需要进行的数据清理包括:
1.（`InvoiceDate`: 发票日期和时间） 变量数据类型转换为日期时间
2.（`CustomerID`: 客户编号） 转换为字符串 
3.（`Description`: 产品名称）变量缺失的观察值删除
4.（`Country`: 国家名称） `USA`变成-->`United States` `UK`,`U.K.`变成-->`United Kingdom`
5.（`["Quantity"]<0 并且 ["InvoiceNo"].str[0] != "C" 并且 ["UnitPrice"] == 0.0`） 删除
6.（``UnitPrice`: 单价`）小于0的删除

```python
# 为了区分开经过清理后数据和原始数据.我们创建新的变量`cleaned_data`,让它为`original_data`复制出的副本。
# 我们之后的清理步骤都将被运用在`cleaned_data`上.
cleand_data = orighinal_data.loc[:,:]
```

```python
# 1.（`InvoiceDate`: 发票日期和时间） 变量数据类型转换为日期时间
# 使用 `pandas.to_datatime("需要转换日期时间的列")` 函数转换。此函数不会产生新数据。
cleand_data["InvoiceDate"] = pd.to_datetime(cleand_data["InvoiceDate"])
cleand_data
```

```python
# 2.（`CustomerID`: 客户编号） 转换为字符串
cleand_data["CustomerID"] = cleand_data["CustomerID"].astype(str)

# 把字符串`17850.0` 的 `.0`删除
# 字符串 str属性下有个 slice方法,可以让只保留字符串的一部分
cleand_data["CustomerID"] = cleand_data["CustomerID"].str.slice(0,-2)   # .str使用字符串属性(这样就能获得很多字符串的方法) .slice()方法参数 == range()参数的逻辑
```

```python
cleand_data
```

```python
# 3.（`Description`: 产品名称）变量缺失的观察值删除
cleand_data = cleand_data.dropna(subset=["Description"])
```

```python
# 打印不为空
cleand_data[cleand_data["Description"].notnull()]
# 查看"Description"为空值的和
cleand_data["Description"].isnull().sum()
```

```python
# 4.（`Country`: 国家名称） `USA`变成-->`United States` `UK`,`U.K.`变成-->`United Kingdom`

#  `USA`变成-->`United States`
#一: 先定位所有 `USA`值
#二: 用.loc[] 方法定位从行开始 定位"Country"列下所有"USA"元素值 并 替换成" United States"
cleand_data.loc[cleand_data["Country"] == "USA","Country"] = "United States"
#三: 查看
cleand_data[cleand_data["Country"] == "United States"]
len(cleand_data[cleand_data["Country"] == "USA"])


# 修改 `UK`,`U.K.`变成-->`United Kingdom`
cleand_data["Country"].replace({"UK":"United Kingdom","U.K.":"United Kingdom"})
cleand_data.loc[(cleand_data["Country"] == "UK") | (cleand_data["Country"] == "U.K."),"Country"] = "United Kingdom"
# 查看
len(cleand_data[(cleand_data["Country"] == "UK") | (cleand_data["Country"] == "U.K.")])
```

```python
# 重新打印 "Country" 列 的每行数据
cleand_data["Country"].value_counts()
```

```python
# 5.（`["Quantity"]<0 并且 ["InvoiceNo"].str[0] != "C" 并且 ["UnitPrice"] == 0.0`） 删除
_ = cleand_data[(cleand_data["Quantity"] < 0) & (cleand_data["InvoiceNo"].str[0] == "C") & (cleand_data["UnitPrice"] == 0.0)].index
cleand_data = cleand_data.drop(_)
```

```python
# 查看
len(cleand_data[(cleand_data["Quantity"] < 0) & (cleand_data["InvoiceNo"].str[0] == "C") & (cleand_data["UnitPrice"] == 0.0)])
```

```python
# 6.（`UnitPrice`: 单价`）小于0的删除  `.index`保存行索引
_ = cleand_data[cleand_data["UnitPrice"] < 0].index
# 把删除赋值
cleand_data = cleand_data.drop(_)
```

```python
# 查看
len(cleand_data[cleand_data["UnitPrice"] < 0])
```

###### 其实删除某些东西还有一种思路，比如把满足某些条件的数据删除掉(我们可以把不满足这些删除条件的数据保存并使用)，就是变相的删除了不满足条件的数据。
在赛选数据前面加上 取反符号即可 ~
`~((cleand_data["Quantity"] < 0) & (cleand_data["InvoiceNo"].str[0] == "C") & (cleand_data["UnitPrice"] == 0.0))`
